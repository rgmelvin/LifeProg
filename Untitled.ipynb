```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kappa_distribution(v, kappa, v_th):
    return(1 / (np.pi * kappa * v_th**2) ** (3/3)) * (1 + (v**2) / (kappa * v_th**2))**(-(kappa + 1))
```

```python
kappa = 3.5
v_th = 1.0
```

```python
v = np.linspace(0, 5, 500)
```

```python
f_v = kappa_distribution(v, kappa, v_th)
```

```python
plt.plot(v, f_v, label=f'k = {kappa}')
plt.xlabel('Velocity')
plt.ylabel('Distribution function')
plt.title('Kappa Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.constants import k, m_u
```

```python
def maxwell_boltzmann_distribution(v, T, m):
    return (m / (2 * np.pi * k * T))**(3 / 2) * 4 * np.pi * v**2 * np.exp(-m * v**2 / (2* k * T))
```

```python
T = 300
m = 40 * m_u
```

```python
v = np.linspace(0, 2000, 500)
```

```python
f_v = maxwell_boltzmann_distribution(v, T, m)
```

```python
plt.plot(v, f_v, label=f'T = {T} K')
plt.xlabel('Velocity (m/s)')
plt.ylabel('Distribution function')
plt.title('Maxwell-Boltzmann Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
